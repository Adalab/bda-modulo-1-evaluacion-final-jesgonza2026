```python

#Definimos la clase con sus respectivos atributos:

class Tienda0nline:
    def __init__(self):
        self.inventario=[] #Creamos una lista vacía ,más adelante guardaremos diccionarios en esta lista.
        self.ventas_totales=0.0 


#1.
#CREAMOS MÉTODO AGREGAR_PRODUCTO:

    def agregar_producto(self,nombre,precio,cantidad): #defino el método
        for producto in self.inventario:  #iteramos por la lista self.inventario con un for.Recorre cada producto en la lista self.inventario
            if producto ["nombre"]==nombre:#revisamos si ya tenemos un producto con ese nombre
                producto["cantidad"]=producto["cantidad"]+cantidad #le decimos: actualiza la cantidad: a la cantidad que ya tiene el producto, súmale esta nueva cantidad.”

                return #si se cumple esa condición ya no queremos no necesitamos que añada ningún producto
                
        self.inventario.append({"nombre":nombre,"precio":precio,"cantidad":cantidad})#al no existir,le indicamos que lo añada
          
#2.
#CREAMOS METODO VER_INVENTARIO
    def ver_inventario(self):
        for producto in self.inventario: ##iteramos por la lista self.inventario con un for.Recorre cada producto en la lista self.inventario
            print("Nombre :" ,producto ["nombre"],", "
                  "Precio: $",producto ["precio"],", "
                  "Cantidad:", producto["cantidad"])
            
#3.
# CREAMOS EL METODO BUSCAR_PRODUCTO:

    def buscar_producto(self,nombre):
        for producto in self.inventario: #iteramos por la lista self.inventario con un for.Recorre cada producto en la lista self.inventario
            if producto["nombre"] == nombre: #si el nombre del producto que tengo coincide con el nombre del producto que me han facilitado
                print("Nombre :" ,producto ["nombre"],", "
                  "Precio: $",producto ["precio"],", "
                  "Cantidad:", producto["cantidad"])
        
                return   #return para que en el momento que encuentre el producto la función se detenga y no siga buscando
#4.
#CREAMOS EL METODO ACTUALIZAR_STOCK:
    def actualizar_stock(self,nombre,cantidad):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                producto["cantidad"]=producto["cantidad"] + cantidad
                
                return
        
        print("el producto no está en el inventario")
    
#5.
#CREAMOS EL METODO ELIMINAR_PRODUCTO

    def eliminar_producto(self,nombre):
        for producto in self.inventario:
           if producto["nombre"] == nombre:
            self.inventario.remove(producto)

            return
        
        print("el producto no está en el inventario")

#6.
#CREAMOS EL METODO CALCULAR_VALOR_INVENTARIO:

    def calcular_valor_inventario(self):

        valor_total_inventario = 0 #creo un acumulador que no es lo mismo que un contador

        for producto in self.inventario:
            valor_producto = producto["precio"] * producto["cantidad"]
            valor_total_inventario=valor_total_inventario + valor_producto

            #aquí no ponemos return xq sino en la primera iteración termina,solo procesará el primer producto
        
        print(valor_total_inventario)
    
```
